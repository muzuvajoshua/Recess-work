```python
#create model
import numpy
import matplotlib.pyplot as plt
numpy.random.seed(2)
x= numpy.random.normal(3,1,100)
y=numpy.random.normal(150,40,100)/x

plt.scatter(x,y)
plt.show()
#its used for generating pseudo random numbers
#the seed is like a starter value for the function based on which the next random number will be generated

#its called pseudo because if somebody has both the algorithm and the seed you used for getting the random number, they can guess it.

#truly random numbers can't be guessed
```

```python
#slit into Train/Test
#Training 80%
#testing 20%

train_x = x[:80]
train_y = y[:80]

test_x = x[:20]
test_y = y[:20]

```
