```python
#web scrapping
import requests
url = 'https://www.xeno-canto.org/api/2/recordings?query=cnt:uganda'
headers = {
    'Accept':'application/json',
    'Content-Type':'application/json'
}
response = requests.request('GET', url,headers=headers, data={})
myjson = response.json()
import pandas as pd
recordings = myjson['recordings']
recording = pd.DataFrame(data=recordings)
recording.to_csv('recrding_UG.csv',index=False)
```

```python
#load the web page conent
#displaying a table
import requests
from bs4 import BeautifulSoup
import pandas as pd


table=soup.select('table.hockey-stats')[0]
columns=table.find('thead').find_all('th')
columns_names=[c.string for c in columns]
table_rows = table.find('tbody').find_all('tr')
l = []
for tr in table_rows:
    td =tr.find_all('td')
    row = [str(tr.get_text()).strip()for tr in td]
    l.append(row)
    
df = pd.DataFrame(l,columns=columns_names)
df.head()
```
